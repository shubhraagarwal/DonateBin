```python
import numpy as np
import pandas as pd
import os
```

```python
# generate labels

t_bed = 'furniture/train/bed'
t_chair = 'furniture/train/chair'
t_sofa = 'furniture/train/sofa'
t_sw = 'furniture/train/sw'
t_table = 'furniture/train/table'

v_bed = 'furniture/valid/bed'
v_chair = 'furniture/valid/chair'
v_sofa = 'furniture/valid/sofa'
v_sw = 'furniture/valid/sw'
v_table = 'furniture/valid/table'

phone = 'mobiles'
fashion = 'fashion'
```

```python
def df_Func(directory, label, dtype):
    img_id = []
    img_label = []
    for i, filename in enumerate(os.listdir(directory)):
        name = label + dtype + '_' + str(i) + '.jpg'
        og_path = os.path.join(directory, filename)
        f_path = os.path.join(directory, name)
        os.rename(og_path,f_path) 
        
        img_id.append(name)
        img_label.append(label)

    df = pd.DataFrame(columns = ['id','label'])
    df['id'] = img_id
    df['label'] = img_label
    return df
        
```

```python

bed_df = df_Func(t_bed,'bed','t')
```

```python
chair_df = df_Func(t_chair,'chair','t')
```

```python
sofa_df = df_Func(t_sofa,'sofa','t')
```

```python
sw_df = df_Func(t_sw,'sw','t')
```

```python
tabel_df = df_Func(t_table,'table','t')
```

```python
bed_df2 = df_Func(v_bed,'bed', 'v')
```

```python
chair_df2 = df_Func(v_chair,'chair', 'v')
```

```python
sofa_df2 = df_Func(v_sofa,'sofa', 'v')
```

```python
sw_df2 = df_Func(v_sw,'sw', 'v')
```

```python
tabel_df2 = df_Func(v_table,'table', 'v')
```

```python
bed_df
```

```python
dataset = pd.concat([bed_df, bed_df2, chair_df, chair_df2, sofa_df, sofa_df2, sw_df, sw_df2,tabel_df, tabel_df2], axis = 0)
```

```python
dataset = dataset.reset_index(drop = True)
dataset
```
